```python
# -*- encoding:utf-8 -*-
import numpy as np
import matplotlib.pyplot as plt
a1 = np.loadtxt("./y.txt")
y1 = a1[:,0]
y2 = a1[:,1]
y3 = a1[:,2]
x = np.linspace(-1,1,len(y1))
y = -10/np.pi*np.cos(np.pi/2*x)
y_1 = 5*np.sin(np.pi/2*x)
y_2 = np.ones((len(y1),))*(np.pi/2)**2
fig = plt.figure(dpi=250)
plt.plot(x,y1,linewidth=1.9,label='Fortran')
plt.plot(x,y,linewidth=1.2,label='Theorem')
plt.legend()
plt.title(r'$\psi-x$')
plt.xlabel('x')
plt.ylabel(r'$\psi$')
plt.grid()
plt.tight_layout()
plt.savefig('./inf1-1.png',dpi=250)
plt.show()

fig1 = plt.figure(dpi=250)
plt.plot(x,y2,linewidth=1.9,label='Fortran')
plt.plot(x,y_1,linewidth=1.2,label='Theorem')
plt.legend()
plt.grid()
plt.title(r'$\psi^{\prime}-x$')
plt.xlabel('x')
plt.tight_layout()
plt.savefig('./inf1-2.png',dpi=250)
plt.show()

fig2 = plt.figure(dpi=250)
plt.plot(x,y3,linewidth=1.9,label='Fortran')
plt.plot(x,y_2,linewidth=1.2,label='Theorem')
plt.legend()
plt.grid()
plt.title(r'$k^{2}-x$')
plt.xlabel('x')
plt.ylabel(r'$k^{2}$')
plt.tight_layout()
plt.savefig('./inf1-3.png',dpi=250)
plt.show()

```

```python
# -*- encoding:utf-8 -*-
import numpy as np
import matplotlib.pyplot as plt
a0 = np.loadtxt("./y0.txt")
a1 = np.loadtxt("./y.txt")
y10 = a0[:,0]
y20 = a0[:,1]
y30 = a0[:,2]
y11 = a1[:,0]
y21 = a1[:,1]
y31 = a1[:,2]
x = np.linspace(-1,1,len(y10))
y12 = -10/np.pi*np.cos(np.pi/2*x)
y22 = 5*np.sin(np.pi/2*x)
y32 = np.ones((len(y10),))*(np.pi/2)**2
plt.rcParams['font.sans-serif']=['SimSun']
plt.rcParams['axes.unicode_minus'] = False

fig, axes = plt.subplots(2, 1, dpi=250,constrained_layout=True)
axes[0].plot(x,y10,linewidth=1.2,color='black')
axes[0].set_title(r'$\psi$'+'')
axes[0].set_xlabel(r'$x$')
axes[0].grid()
axes[1].plot(x,y11,linewidth=2.9,color='black',linestyle=':',label='Fortran')
axes[1].plot(x,y12,linewidth=1.2,color='black',linestyle='-',label='')
axes[1].set_title(r'$\psi$')
axes[1].set_xlabel(r'$x$')
axes[1].legend()
axes[1].grid()
fig.savefig('./inf1-psi1.png',dpi=250)

fig, axes = plt.subplots(2, 1, dpi=250,constrained_layout=True)
axes[0].plot(x,y20,linewidth=1.2,color='black')
axes[0].set_title(r'$\psi^{\prime}$'+'')
axes[0].set_xlabel(r'$x$')
axes[0].grid()
axes[1].plot(x,y21,linewidth=2.9,color='black',linestyle=':',label='Fortran')
axes[1].plot(x,y22,linewidth=1.2,color='black',linestyle='-',label='')
axes[1].set_title(r'$\psi^{\prime}$')
axes[1].set_xlabel(r'$x$')
axes[1].legend()
axes[1].grid()
fig.savefig('./inf1-psi2.png',dpi=250)

fig, axes = plt.subplots(2, 1, dpi=250,constrained_layout=True)
axes[0].plot(x,y30,linewidth=1.2,color='black')
axes[0].set_title(r'$k^{2}$'+'')
axes[0].set_xlabel(r'$x$')
axes[0].grid()
axes[1].plot(x,y31,linewidth=2.9,color='black',linestyle=':',label='Fortran')
axes[1].plot(x,y32,linewidth=1.2,color='black',linestyle='-',label='')
axes[1].set_title(r'$k^{2}$')
axes[1].set_xlabel(r'$x$')
axes[1].legend(loc='right')
axes[1].grid()
fig.savefig('./inf1-psi3.png',dpi=250)
plt.show()
```
